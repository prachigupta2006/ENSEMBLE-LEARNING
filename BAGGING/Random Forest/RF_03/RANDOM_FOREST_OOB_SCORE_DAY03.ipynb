### **GRIDSEARCHCV**

```python
# Number of trees in the random forest
n_estimators = [10 ,20 ,40, 60]

# Number of features to consider at every split
max_features  = [0.2 ,0.4, 0.5 , 0.8 , 1.0]

# Maximum number of levels in tree
max_depth  = [2,4,6,8]

# Minimum number of samples required
min_samples_split = [0.5,0.60 , 0.75, 1.0]
```

```python
param_grid  = {
              'n_estimators' : n_estimators,
              'max_features' : max_features,
              'max_depth' : max_depth,
              'min_samples_split' : min_samples_split
}
print (param_grid)

rf = RandomForestClassifier()
```

```python
rf_grid= GridSearchCV(estimator=rf,
                      param_grid=param_grid,
                      cv=5,
                      verbose=2,
                      n_jobs=-1)
```

```python
rf_grid.fit(x_train,y_train)
```

```python
rf_grid.best_params_
```

```python
rf_grid.best_score_
```

### **RANDOM SEARCH CV**

```python
n_estimators = [20 , 40 , 60]

max_features = [0.2 , 0.3 , 0.5 , 0.6 ]

max_depth  = [ 2 , 4 , 6 , 8 ]

max_samples = [0.5 ,0.65 ,  0.75]

bootstrap = [True , False]

min_samples_split = [2 , 4  ,  6]

min_samples_leaf = [ 2 , 3 , 4 ]
```

```python
param_grid = {
              'n_estimators' : n_estimators,
              'max_features' : max_features,
              'max_depth' : max_depth,
              'max_samples' : max_samples,
              'bootstrap' : bootstrap,
              'min_samples_split' : min_samples_split,
              'min_samples_leaf' : min_samples_leaf

}
print (param_grid)
```

```python
rf = RandomForestClassifier()

from sklearn.model_selection import RandomizedSearchCV

rf_grid  = RandomizedSearchCV(estimator=rf , param_distributions=param_grid , cv=5 , verbose=2 , n_jobs=-1)
```

```python
rf_grid.fit(x_train,y_train)
```

```python
rf_grid.best_params_
```

```python
rf_grid.best_score_
```
